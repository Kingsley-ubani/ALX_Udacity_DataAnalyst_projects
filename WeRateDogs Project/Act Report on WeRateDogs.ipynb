### Insights generated from WeRateDogs project

This report describes the insights and  visualization gotten from the WeRateDogs Project.



#### Insight 1 - Top 10 most tweeted dog_breed

The visuals below shows the top 10 most engaged Dog Breed by tweet.





From the analysis on the WeRateDogs cleaned data shared in the wrangle_act file, the total number of tweets gotten was 2075 tweets. However, for the sake of validation,it is safe to say golden_retriever is the most tweeted dog-breed (having 173 tweets). This is because the top of the chart is an 'unknown' breed which can't be most likely assigned to any breed on the list. The top 5 most tweeted dog-breed are Golden_retriever,Labrador_retriever,Pembroke, Chihuahua and Pug.



#### Insight 2 - Dog_breed with the most rating

The visuals below describes the top 10 Dog-breed with the most rating 



From the graph, it can be clearly seen that the top 5 dog-breed with the most rating are Golden_retriever,Labrador_retriever,Pembroke,Chihuahua and Pug. Golden retriever retains it's position as the dog breed with most rating.

This also didn't come as a surprise because the same list is the most tweeted. From findings,it is worthy of note to conclude that these breeds are friendly and make great family pets.
